## Preprocessing

```python
air2022.drop(columns = ["데이터기준일자"], inplace = True)
air2022.head(5)
```

```python
air2022.rename(columns = {"폼알데하이드" : "포름알데히드"}, inplace = True)
air2022.head(5)
```

```python
air2022.replace({"역명" : 
               {"서울역(1)" : "서울역"}},
               inplace = True)
air2022.replace({"역명" : 
               {"시청(1)" : "시청"}},
               inplace = True)
air2022.head(5)
```

```python
air2021.drop(columns = ["이산화질소", "휘발성유기화합물", "라돈"], inplace = True)
air2021.head(5)
```

```python
air2019.drop(columns = ["이산화질소", "휘발성유기화합물", "라돈"], inplace = True)
air2019.head(5)
```

```python
air2018["연번"] = range(1, len(air2018) + 1)
air2018.set_index("연번", inplace = True)
air2018.head(5)
```

```python
air2018.rename(columns = {"미세먼지_PM-10(140 ㎍/㎥ 이하)" : "미세먼지", "이산화탄소_CO2 (1000 ppm 이하)" : "이산화탄소", "포름알데히드_HCHO(100 ㎍/㎥ 이하)" : "포름알데히드", "일산화탄소_CO(9 ppm이하)" : "일산화탄소"}, inplace = True)
air2018.head(5)
```

## Visualization

```python
air2022_1 = air2022[air2022["호선"] == "1호선"]
air2022_1.head(5)
```

```python
sns.boxplot(y = "미세먼지", data = air2022_1)
plt.title("2022년 1호선 미세먼지 분포도")
plt.show()
```

```python
air2022_2 = air2022[(air2022['호선'] == '2호선') & (air2022['역명'] == '신촌')]
air2022_2.head(5)
```

```python
air2021_2 = air2021[(air2021['호선'] == 2) & (air2021['역명'] == '신촌')]
air2021_2.head(5)
```

```python
air2020_2 = air2020[(air2020['호선'] == 2) & (air2020['역명'] == '신촌')]
air2020_2.head(5)
```

```python
form3year = pd.concat([air2022_2, air2021_2, air2020_2], axis=0)
form3year.head(5)
```

```python
sns.barplot(x = "연도", y = "포름알데히드", data = form3year, palette = "Greens")
plt.title("최근 3년간 2호선 신촌역 포름알데히드 검출량")
plt.show()
```

```python
air2022_3 = air2022[air2022["호선"] == "3호선"]
air2022_3 = air2022_3.loc[:, ["미세먼지", "초미세먼지", "역명"]]
print(air2022_3)
```

```python
sns.scatterplot(x = "미세먼지", y = "초미세먼지", data = air2022_3, hue = "역명")
plt.title("미세먼지와 초미세먼지의 관계, 2022년 3호선")
plt.show()
```

```python
sns.kdeplot(data = air2022, x = "일산화탄소")
sns.rugplot(data = air2022, x = "일산화탄소")
plt.title("2022년 지하철역 일산화탄소 분포")
plt.show()
```

```python
air2021_5 = air2021[air2021["역명"] == "동대문역사문화공원"]
print(air2021_5)
```

```python
air2020_5 = air2020[air2020["역명"] == "동대문역사문화공원"]
print(air2020_5)
```

```python
air2019_5 = air2019[air2019["역명"] == "동대문역사문화공원"]
print(air2019_5)
```

```python
co2year3 = pd.concat([air2021_5, air2020_5, air2019_5], axis=0)
print(co2year3)
```

```python
sns.lineplot(x = "연도", y = "이산화탄소", data = co2year3, hue = "호선")
plt.xticks([2019, 2020, 2021])
plt.title("동대문역사문화공원 이산화탄소 변화율 (2019~2021)")
plt.plot()
```

```python
air2020_6 = air2020[air2020["호선"] == 4]
print(air2020_6)
```

```python
air2020_6.set_index("역명", inplace = True)
air2020_6.drop(columns = ["호선", "연도"], inplace = True)
print(air2020_6)
```

```python
sns.heatmap(air2020_6, annot = True, cmap = "YlGnBu")
plt.title("2020년 서울 지하철 4호선 공기지수 히트맵")
plt.show()
```

```python
sns.violinplot(data = air2020_6, palette = "Set2")
plt.title("2020년 서울 지하철 4호선 공기지수 바이올린플롯")
plt.show()
```
